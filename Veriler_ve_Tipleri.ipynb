```python
###Burda verileri tekrardan görecez ve veri tipini ekrana yazdıracaz

#Bool (var/yok,1 veya 0)
a = True
print(type(a))

#integer (int)
a = 1
print(type(a))

#string (str)
a ="Gökhan"
print(type(a))

#double (float)
a=4.8
print(type(a))
```

```python
###Burda verileri dönüştürme işlemi var
### aslında biraz değişik bir sisteme sahip olsada bütün olayı
###yazı olarak girilmiş sayı varsa int e çevirir yada diğer sayısal
###verilere yada sayıları yazı yapar işte

#int > str
a = 172
b= str(a)
print(b)
```

```python
#str > int
a = "123"
b = int(a)
print(b)
```

```python
#str > float(namı diğer double türü)
a ="123.01"
b = float(a)
print(b)
```
